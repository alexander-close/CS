```python
import pandas as pd

raw_data=pd.read_csv('Billionaires Statistics Dataset.csv')
df=pd.DataFrame(raw_data)

df.describe()
```

```python
#Check for null values throghouut the dataset

for col in df.columns:
    NaNs = sum(df[col].isnull())
    if NaNs != 0:
        print(f'There are {NaNs} null values in column {col}.')
print('The other columns do not have null values.')
```

```python
#function that replaces null ages etc. with median age etc. 
#then converts to integer format

def null_int_med(df,col):
    df[df[col].isnull()] = df[df[col].notnull()][col].median()
    return df[col].apply(int)
    

df['age'] = null_int_med(df,'age')
df['birthYear'] = null_int_med(df,'birthYear')
df['birthMonth'] = null_int_med(df,'birthMonth')
df['birthDay'] = null_int_med(df,'birthDay')

df[['age','birthYear','birthMonth','birthDay']]
topten=df.sort_values(['age']).tail(10)
topten
```

```python
def some_decorator(f):
    def wraps(*args):
        print(f"Calling function '{f.__name__}'")
        return f(args)
    return wraps

@some_decorator
def decorated_function(x):
    print(f"With argument '{x}'")
```

```python
def average_age(members, gender):
   
  total_age = 0
  count = 0
   
  for member in members:
    name, age, member_gender = member
    
    if member_gender == gender:
      total_age += age
      count += 1
   
  if count > 0:
    average = total_age / count  
    return (average)
  else:
    return ('No matches found.')
```
